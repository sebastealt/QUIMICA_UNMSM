```python
# Importamos
from drawdata import draw_scatter, draw_histogram, draw_line
# seleccionar A y graficar  la distribución
draw_scatter()
```

```python
df = pd.read_csv("data (3).csv")
df
```

```python
df['x'] = 1/(df['x'])
df['x']
```

```python
plt.figure(figsize=(200,10))
plt.scatter(df['x'],df['y'])
plt.xlabel('1/VOLUMEN')
plt.ylabel('PRESION')
plt.title('Ley de BOYLE')
```

```python
volumen = df["x"].values.reshape(-1, 1)
presion = df["y"].values
modelo = LinearRegression()


modelo.fit(volumen, presion)

k = modelo.coef_[0]

b = modelo.intercept_
print(f"Ecuación de la Ley de Boyle: P = {k} * (1/V) + {b}")
```
